```python
#more stuff
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('darkgrid')
tipgraph = sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip',
             color = 'lime',
             linewidth = 1,
             marker = 'o',
             markerfacecolor = 'maroon')
tipgraph.set(title = "tips vs total bill", 
             xlabel = 'total bill ($$$)', 
             ylabel = 'tip amount ($$$)')
```

```python
tipgraph = sns.scatterplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        style = 'time',
                        hue = 'day', palette = ['pink', 'purple', 'gold', 'skyblue'])
```

```python
#do pairplots
sns.pairplot(data = tips, hue = 'day')
#essentially figures out what the combinations are, quantitative against each other
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'day', kind = 'scatter', col_wrap = 2, hue = 'day', palette = ['aquamarine', 'rosybrown'])
```

```python
import plotly.graph_objects as go
x1 = np.array(["peas", "corn", "carrots", "potatoes"])
y1 = np.array([12.7, 19.5, 9.8, 1.2])
graph = go.Figure()
```

```python
graph.add_trace(go.Bar(x = x1, y = y1))
```

```python
graph.update_layout(title = 'veggie mix', 
                       xaxis_title = 'veg',
                       yaxis_title = 'annoyingness')
#oh interesting, it is interactive
#can save as html
graph.write_html("veggie_mix.html")
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x = x1, y = y1, 
                           mode = 'markers', 
                           marker = dict(size = 15, 
                                         color = 'rebeccapurple',line = dict(width = 2, color = 'bisque'))))

graph.update_layout(title = 'veggie mix', 
                       xaxis_title = 'veg',
                       yaxis_title = 'annoyingness', width = 500, height = 500)
```

```python
#code for qualitative information -> still reproducible!
from wordcloud import WordCloud
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/refs/heads/master/animal_crossing/fish.csv')
df
```

```python
#concat
text = " ".join(each for each in df.Name)
#generate cloud
wordcloud = WordCloud(background_color='paleturquoise', colormap = 'mako').generate(text)
fig,ax = plt.subplots(figsize = (7,3))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2_unweighted

A = set(['butterfly', 'worm', 'beetle', 'isopod'])
B = set(['isopod', 'lobster', 'squat crab', 'crayfish'])

#venn diagram
diagram = venn2_unweighted([A,B], 
                           set_labels = ('bugs', 'shrimp'), 
                           set_colors = ('forestgreen', 'salmon'), 
                           alpha = 0.5)

diagram.get_label_by_id("10").set_text("\n".join(A-B))
diagram.get_label_by_id("11").set_text("\n".join(A&B))
diagram.get_label_by_id("01").set_text("\n".join(B-A))
```
