```python
fig,bee=plt.subplots(figsize=(5,3))
#the next line determines what kind of plot is drawn
bee.plot(x,y)
```

```python
fig,bee=plt.subplots(figsize=(5,3))
bee.set_title('how much do bees go buzz?')
bee.set_ylabel('buzz frequency')
bee.set_ylabel('buzz volume (dB)')
#the next line determines what kind of plot is drawn
#changing some 'artists'
bee.hist(y, rwidth=0.8, edgecolor = 'black', color = 'green')
fig.tight_layout()
```

```python
font1={'family':'Comic Sans MS', 'color':'purple', 'size':15}
font2={'family':'papyrus', 'color': '#2D6FE3','size':10}
```

```python
fig,bee=plt.subplots(figsize=(5,3))
bee.set_title('how much do bees go buzz?', fontdict = font1)
bee.set_ylabel('buzz frequency', fontdict = font2)
bee.set_xlabel('buzz volume (dB)', fontdict = font2)
bee.hist(y, rwidth=0.8, edgecolor = 'black', color = 'green')
```

```python
fig,bee=plt.subplots(figsize=(5,3))
bee.set_title('how much do bees go buzz?', fontdict = font1)
bee.set_ylabel('buzz frequency', fontdict = font2)
bee.set_xlabel('buzz volume (dB)', fontdict = font2)
bee.scatter(x,y, marker = '1', color = 'teal')
```

```python
fig,bee=plt.subplots(figsize=(5,3))
bee.set_title('how much do bees go buzz?', fontdict = font1)
bee.set_ylabel('buzz frequency', fontdict = font2)
bee.set_xlabel('buzz volume (dB)', fontdict = font2)
bee.plot(x,y, marker = '*', color = 'teal', linestyle = ':', markerfacecolor = '#DE0DBB', markeredgecolor = '#1ED4E6')
```

```python
fig,bee=plt.subplots(figsize=(5,3))
bee.set_title('how much do bees go buzz?', fontdict = font1)
bee.set_ylabel('buzz frequency', fontdict = font2)
bee.set_xlabel('buzz volume (dB)', fontdict = font2)
bee.grid(axis='y', color = '#EDAA3E')
bee.grid(axis='x', color = '#49D10A')
bee.plot(x,y, marker = 'X', color = 'teal', linestyle = ':')
```
