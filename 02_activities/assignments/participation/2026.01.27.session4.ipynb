```python
#new data viz slides, subplots and stuff
np.random.seed(833833)
plt.style.use('seaborn-v0_8-deep')
x = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(['isopod', 'sea slug', 'worm', 'shrimp', 'snail'])
y2 = np.array([110, 180, 240, 99, 220])



fig, (ax1, ax2) = plt.subplots(ncols =1, nrows =2, figsize = (3, 5))
ax1.scatter(x, y1, color = 'teal')
ax2.bar(x2, y2)
```

```python
x3 =  np.random.randint(0,25,5)
y3 = np.array(['dirt', 'mud', 'leaves', 'water', 'goop'])

fig, mult_axes = plt.subplot_mosaic([['ax1', 'ax3'],
                                     ['ax2', 'ax3']], figsize = (8,8))
mult_axes["ax1"].scatter(x, y1, s = 1)
mult_axes["ax1"].set_xlabel('what ants look like from really far away')


mult_axes["ax2"].bar(x2, y2)
mult_axes["ax2"].set_xlabel('sliminess')


mult_axes["ax3"].bar(y3, x3)
mult_axes["ax3"].set_xlabel('habitat quality')


```

```python
y2_sd = np.std(y2)
fig, ax=plt.subplots(figsize = (7,3))
# ax.plot(x2, y2, color = 'lime')
ax.errorbar(x2, y2, ecolor = 'darkgreen', color = 'indigo', fmt = '-', yerr= y2_sd, capsize = 10, capthick = 4)
```
