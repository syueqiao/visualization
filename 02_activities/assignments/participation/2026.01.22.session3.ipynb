```python
#using comments in code :D
#part of assignment 1?
#overview of plot adjustment and changes
np.random.seed(833833)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python

#specify information such as size
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'wahoo')
ax.plot(x,y2, label = 'yipee')
ax.legend(loc= 'best')
```

```python

#specify information such as size
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = 'wahoo')
ax.plot(x,y2, label = 'yipee')
#mess around with the aesthetics
ax.legend(loc= 'best', frameon = True, shadow = True, bbox_to_anchor = (1,1))
```

```python

#specify information such as size
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa')

ax.text(4, 95, 'beep boop', color = 'pink')
#do not need fig.show for jupyter notebook use
#can also change with (ha) to modify the horizontal alignment
```

```python

#specify information such as size
fig, ax = plt.subplots()
ax.axis([0,10,0,10])
#additional positioning of annotation to other parts of the plot, eg axis
ax.text(1,5, '.Data:(1,5)', transform = ax.transData)
ax.text(0.5,0.5, '.Axes:(0.5,0.5)', transform = ax.transAxes)
ax.text(0.2,0.5, '.Figure:(0.2,0.5)', transform = fig.transFigure)
#what is the relationship between the first and second argument?
```

```python

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa')

ax.text(4, 95, 'beep boop', color = 'pink')
ax.annotate('wow', xy = (10,85), xytext = (30, 94), arrowprops=dict(facecolor = 'green'))
```

```python

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa')

ax.text(4, 95, 'beep boop', color = 'pink')
ax.annotate('wow!!!!', xy = (10,85), xytext = (30, 94), arrowprops=dict(arrowstyle = 'wedge', facecolor = 'lime'))
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa')
ax.legend(loc= 'best', frameon = True, shadow = True, bbox_to_anchor = (1,1))
ax.yaxis.set_major_locator(plt.NullLocator())
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa', color = 'green')
ax.legend(loc= 'best', frameon = True, shadow = True, bbox_to_anchor = (1,1))
ax.yaxis.set_major_locator(plt.MaxNLocator(2))
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa')
ax.legend(loc= 'best', frameon = True, shadow = True, bbox_to_anchor = (1,1))
ax.yaxis.set_major_locator(plt.MultipleLocator(1))
#wahahaha ugly
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = 'yippee')
ax.scatter(x, y2, label = 'wahaaa')
ax.legend(loc= 'best', frameon = True, shadow = True, bbox_to_anchor = (1,1))
plt.xticks(rotation = 45, ha = 'center')
```
